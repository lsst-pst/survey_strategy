```python
# Pull together DDF metric summaries from temporary homes
dd = pd.read_csv('/Users/lynnej/opsim/v2.2/baseline_new/ddf_summary.csv', index_col=0)
dd = dd.join(pd.read_csv('/Users/lynnej/opsim/v2.2/baseline_new/meta_summary.csv', index_col=0))
dd = dd.join(pd.read_csv('/Users/lynnej/opsim/v2.2/baseline_new/sci_summary.csv', index_col=0))
tt = pd.read_csv('summary_2022_08_01.csv', index_col=0).query('run_name.str.startswith("baseline")', engine='python')
ddf = pd.read_csv('ddf_summaries.csv', index_col=0)
ddf = pd.concat([ddf, tt, dd])
ddf.to_csv('ddf_v2.csv')
```

```python
# Load up specific DDF summary file - contains DDF metrics (only) for the ddf v2.2 and baseline v2.2/v2.1 runs.
# baseline_v2.1 is a non-pre-scheduled DDF run.
# baseline_v2.2 has a new, prescheduled DDF run of an "intermediate" kind, and year 1 WFD coverage fixed.
# ddf_* v2.2 runs are a variety of DDF strategies, although the WFD year 1 u band coverage is incomplete - they 
# should still be valid for interpreting best DDF strategies.

ddf_summaries = maf.get_metric_summaries(summary_source='ddf_v2.csv')
metric_sets = maf.get_metric_sets('metric_sets.json')
```

```python
ddf_summaries.head()
```

```python
# Get a list of the runs available to investigate
all_runs = ddf_summaries.index.to_list()
# But we need to organize them. 

# Get the metric sets for DDF fields
msets = list(metric_sets.groupby('metric set').first().index)
msets = [m for m in msets if 'DDF' in m]
for k in msets:
    print(k, len(metric_sets.loc[k]))
    # Check that the metrics are correctly named and in the summary file
    ddf_summaries.loc[:, metric_sets.loc[k]['metric']]
```

A brief overview of the metric sets, for all DDF simulations. 

```python
baseline = 'baseline_v2.2_10yrs'

these_runs = all_runs

for m in ['DDF Depths', 'DDF Cadence']:
    metrics = metric_sets.loc[m]
    fig, ax = maf.plot_run_metric_mesh(ddf_summaries.loc[these_runs, metrics['metric']],
                                      baseline_run=baseline,
                                      metric_label_map=metrics['short_name'],
                                      metric_set=metrics, 
                                      color_range=1)
    fig.set_figwidth(20)
    fig.set_figheight(12)
```

```python
ddfs = ['COSMOS', 'XMM_LSS', 'ELAISS1', 'ECDFS', 'EDFS']

fig = plt.figure(figsize=(15, 15))
gs = fig.add_gridspec(ncols=5, nrows=1, wspace=0.1)
axs = gs.subplots(sharey=True)
for dd, ax  in zip(ddfs, axs):
    msubset = metrics.query('metric.str.contains(@dd)')
    fig, ax = maf.plot_run_metric(ddf_summaries.loc[these_runs, msubset['metric']],
                                  baseline_run=baseline,
                                  metric_label_map=msubset['short_name'],
                                  metric_set=msubset,
                                 horizontal_quantity='value',
                                 vertical_quantity='run',
                                 ax=ax)
    ax.get_legend().remove()
    ax.set_xlim(0.4, 1.6)
    ax.set_title(dd)
ax.legend(loc=(1.01, 0.5))
```
